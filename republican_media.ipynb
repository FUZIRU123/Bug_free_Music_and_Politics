# kanye west

```python
cleaned_kw_df1 = kw_df[kw_df['ParentID'].isna()== True]
```

```python
cleaned_kw_df2 = pd.DataFrame({'Comment': cleaned_kw_df1['Comment'],
'User_name': cleaned_kw_df1['AuthorDisplayName'],
'User_channel_id': cleaned_kw_df1['AuthorChannelID'],
'Video_id': cleaned_kw_df1['VideoID']
})
cleaned_kw_df2['Video_author'] = 'Kanye_West'
cleaned_kw_df2.reset_index(drop=True, inplace=True)
cleaned_kw_df2.head(6)
```

```python
for i in range(len(cleaned_kw_df2['Comment'])):
    cleaned_kw_df2['Comment'][i] = str(cleaned_kw_df2['Comment'][i])
    cleaned_kw_df2['Comment'][i] = demoji.replace(cleaned_kw_df2['Comment'][i],"")
cleaned_kw_df2['Comment']
```

```python
cleaned_kw_df2= cleaned_kw_df2[-cleaned_kw_df2.Comment.isin([""])]
cleaned_kw_df2.reset_index(drop=True, inplace=True)
cleaned_kw_df2
```

```python
cleaned_kw_df2.to_csv('../data/cleaned_republican/cleaned_Kanye_West.csv')
```
